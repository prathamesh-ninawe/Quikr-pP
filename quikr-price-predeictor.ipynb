```python
import pandas as pd
```

```python
car = pd.read_csv("quikr_car.csv")
```

```python
car = car[car['year'].str.isnumeric()]
```

```python
car['year'] = car['year'].astype(int)
```

```python
car = car[car['Price']!= 'Ask For Price']
```

```python
car['Price'] = car['Price'].str.replace(',','').astype(int)
```

```python
car['kms_driven']=car['kms_driven'].str.split( ).str.get(0).str.replace(',','')
```

```python
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
car
```

```python
car = car[~car['fuel_type'].isna()]
```

```python
car
```

```python
car['name']=car['name'].str.split().str.slice(start=0,stop=3).str.join(' ')
```

```python
car
```

```python
car = car.reset_index(drop=True)
```

```python
car = car[car['Price']<6e6].reset_index(drop=True)
```

```python
#Model

```

```python
x = car.drop(columns='Price') #features
y = car['Price'] #target
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
ohe  = OneHotEncoder()
ohe.fit(x[['name','company','fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name', 'company', 'fuel_type']),remainder= 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```

```python
scores=[]
for i in range(1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.005,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(x_train,y_train)
    y_pred=pipe.predict(x_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
import numpy as np
np.argmax(scores)
scores[np.argmax(scores)]
```

```python
np.argmax(scores)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.005,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(x_train,y_train)
y_pred=pipe.predict(x_test)
r2_score(y_test, y_pred)
```
